## ***1. Applying the Monte Carlo Method for Pi Estimation***

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

N = 10000
M = 0

XCrircle = []
YCrircle = []
XSquare = []
YSquare = []

for p in range (N):
  x=random.random()
  y=random.random()

  if (x**2+y**2 <= 1):
        M+=1
        XCrircle.append(x)
        YCrircle.append(y)
  else:
        XSquare.append(x)
        YSquare.append(y)

Pi = 4*M/N
print("N=%d M=%d Pi=%.2f" %(N,M,Pi))
```

```python
XLin=np.linspace(0,1)
YLin=[]
for x in XLin:
  YLin.append(math.sqrt(1-x**2))

plt.axis("equal")
plt.grid(which="major")
plt.plot(XLin, YLin, color="black", linewidth="4")
plt.scatter(XCrircle, YCrircle, color="pink", marker=".")
plt.scatter(XSquare, YSquare, color="blue", marker=".")
plt.title("Monte Carlo Method for Pi Estimation")
plt.show()
```

### **2.** ***Using Monte Carlo in Central Limit Theorem***

```python
import random
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = 100
N = 10000

DataPop = list(np.random.uniform(a,b,N))

plt.hist(DataPop, density=True, histtype='stepfilled', alpha=0.2)
plt.show()

SamplesMeans = []
for i in range(0,10000):
  DataExtracted = random.sample(DataPop, k=100)
  DataExtractedMean = np.mean(DataExtracted)
  SamplesMeans.append(DataExtractedMean)

plt.figure()
plt.hist(SamplesMeans, density=True, histtype='stepfilled', alpha=0.2)
plt.show()
```

## ***3. Use Gamma Distribution in #2***

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gamma distribution
k_shape = 2  # shape parameter
theta_scale = 2 # scale parameter (often denoted as theta or beta)
N_population = 10000 # Population size

# Generate population data using Gamma distribution
DataPop_gamma = list(np.random.gamma(k_shape, theta_scale, N_population))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(DataPop_gamma, bins=50, density=True, histtype='stepfilled', alpha=0.7, color='orange')
plt.title('Population Distribution (Gamma)')
plt.xlabel('Value')
plt.ylabel('Density')

# Sample and calculate means (similar to previous section)
num_samples = 10000
sample_size = 100

SamplesMeans_gamma = []
for i in range(num_samples):
  DataExtracted_gamma = random.sample(DataPop_gamma, k=sample_size)
  DataExtractedMean_gamma = np.mean(DataExtracted_gamma)
  SamplesMeans_gamma.append(DataExtractedMean_gamma)

plt.subplot(1, 2, 2)
plt.hist(SamplesMeans_gamma, bins=50, density=True, histtype='stepfilled', alpha=0.7, color='green')
plt.title(f'Distribution of Sample Means (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
