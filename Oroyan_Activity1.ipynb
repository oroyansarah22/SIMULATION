```python
# @title 1. Create an array of 100,000 random values equally distributed between -2 and +3.3. Print the lowest and highest values of this array.
import numpy as np

# Create an array of 100,000 random values equally distributed between -2 and +3.3
random_array = np.random.uniform(low=-2, high=3.3, size=100000)

# Print the lowest and highest values of this array
print(f"Lowest value: {np.min(random_array)}")
print(f"Highest value: {np.max(random_array)}")
```

```python
# @title 2. Create an array of 100,000 values which are each the sum of two different random values. Print the average of the values in this array.
# Create two arrays of 100,000 random values
random_array1 = np.random.rand(100000)
random_array2 = np.random.rand(100000)

# Create a new array where each value is the sum of the corresponding values in the two random arrays
sum_array = random_array1 + random_array2

# Print the average of the values in this new array
print(f"Average of the sum array: {np.mean(sum_array)}")
```

```python
# @title 3. Create an array of 100,000 values which are each the multiplication of two different random values. Print the average of the values in this array.
# Create two arrays of 100,000 random values
random_array1 = np.random.rand(100000)
random_array2 = np.random.rand(100000)

# Create a new array where each value is the product of the corresponding values in the two random arrays
product_array = random_array1 * random_array2

# Print the average of the values in this new array
print(f"Average of the product array: {np.mean(product_array)}")
```

```python
# @title 4. Create an array of 100,000 values using a normal distribution with an average of 0.3 and a standard deviation of 1.2. Print the average and standard deviation of the values in this array.
import numpy as np

# Create an array of 100,000 values from a normal distribution with mean 0.3 and standard deviation 1.2
normal_array = np.random.normal(loc=0.3, scale=1.2, size=100000)

# Print the average and standard deviation of the values in this array
print(f"Average of the normal array: {np.mean(normal_array)}")
print(f"Standard deviation of the normal array: {np.std(normal_array)}")
```

```python
# @title 5. Create two arrays, a and b, where each are 100,000 random integers with values between 0 and 20. Create a new array, c, from c[i] = a[i] - b[i-1]. Of course, the c array will have a length of 99,999. Create a histogram of these values with a range of -20 to +21.
import numpy as np
import matplotlib.pyplot as plt

a = np.random.randint(0, 21, size=100000)
b = np.random.randint(0, 21, size=100000)

# Create a new array c where each value is a[i] - b[i-1]
c = a[1:] - b[:-1]

# Create a histogram of the values in array c
plt.figure(figsize=(10, 6))
plt.hist(c, bins=range(-20, 22), range=(-20, 21), align='left', rwidth=0.8, edgecolor = 'black')
plt.title("Histogram of Array c")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# @title 6. Create an array of 100,000 values from a normal distribution with an average of 3 and a standard deviation of 1.2. Plot the histogram of these values with a range of -0.6 to +0.6 and a bin width of 0.05.
# Create an array of 100,000 values from a normal distribution with mean 3 and standard deviation 1.2
mean = 3
std_dev = 1.2
normal_array = np.random.normal(loc=mean, scale=std_dev, size=100000)

# Calculate the number of bins
hist_range = (-0.6, 0.6)
bin_width = 0.05
bins = np.arange(hist_range[0], hist_range[1] + bin_width, bin_width)

# Create a histogram of the values in the array
plt.hist(normal_array, bins=bins, range=hist_range, edgecolor = 'black')

# Add labels and a title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution Values")
plt.xlim

# Display the histogram
plt.show()
```
